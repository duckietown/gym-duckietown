# Reinforcement Learning Metrics Dashboard

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("~/dev/harvard/gym-duckietown/learning/metrics-training-spinning.csv") 
data.head()
```

```python
data.tail()
```

```python
fields=['x', 'y', 'angle', 'speed', 'steering', 'center_dist', 'center_angle','reward', 'total_reward']

for field in fields:
    plt.figure(figsize=(15, 2))
    sns.lineplot(x='step',y=field, data=data)
    plt.show()
    
```

```python
#Expected: As model improves bot should start tracing the road
#Actual:   Bot is spinning in place or going forward/backward in place
plt.figure(figsize=(15, 5))
sns.set_style("whitegrid")
sns.scatterplot(x="x", y="y", hue='step', data=data)
```

```python
#Expected: As model improves, center distance and center angle should cluster near 0, 0
#Actual: center distance and center angle are wildly scattered
plt.figure(figsize=(15, 5))
sns.scatterplot(x="center_dist", y="center_angle", hue='step', data=data)
```

```python
# Expected: Reward should be increase as center angle approaches 0
# Actual:   There is no correlation between center angle and reward
# Expected: As model improves, center angle should cluster closer to 0
# Actual:   Center angle is clustering between -1.5 and +1.5
plt.figure(figsize=(15, 5))
sns.scatterplot(x="center_angle", y="reward", hue='step', data=data)
```

```python
# Expected: As model improves, center dist should cluster closer to 0
# Actual:   All distances seem equally present, suggesting bot is all over the road
plt.figure(figsize=(15, 5))
sns.scatterplot(x="center_dist", y="reward", hue='step', data=data)
```

```python
plt.figure(figsize=(15, 5))
sns.scatterplot(x="speed", y="angle", hue='step', data=data)
```
