# Training Image Warping Tests

```python
import cv2
from matplotlib import pyplot as plt
import numpy  as np

img0 = cv2.imread("img10.png")
img1 = cv2.imread("img510.png")
img2 = cv2.imread("img1010.png")

img0 = cv2.cvtColor(img0, cv2.COLOR_BGR2RGB)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(img0)
plt.title('img0')
plt.show()
plt.imshow(img1)
plt.title('img1')
plt.show()
plt.imshow(img2)
plt.title('img2')
plt.show()
```

```python
vanish_height = 50
h = img0.shape[0]
w = img0.shape[1]
dw = 40
src = np.float32([
    [((w - 20) /2) - dw, vanish_height], [((w  + 20) / 2) + dw, vanish_height],
    [0 - dw,            h]           , [w + dw,              h]])
dst = np.float32([[0, 0], [w, 0], [0, h], [w, h]])

#img0 = cv2.line(img0, (src[0][0],src[0][1]), (src[1][0],src[1][1]), (255,0,128), 2) 
#img0 = cv2.line(img0, (src[0][0],src[0][1]), (src[2][0],src[2][1]), (255,128,128), 2) 
#img0 = cv2.line(img0, (src[1][0],src[1][1]), (src[3][0],src[3][1]), (255,128,0), 2) 
#img0 = cv2.line(img0, (src[2][0],src[2][1]), (src[3][0],src[3][1]), (255,128,128), 2) 

M = cv2.getPerspectiveTransform(src, dst)
img_size = (img0.shape[1], img0.shape[0])
warped = cv2.warpPerspective(img0, M, img_size)


plt.imshow(img0)
plt.title('img0')
plt.show()
plt.imshow(warped)
plt.title('warped')
plt.show()
```
